### Fibonacci retracements

```python
min_value = stock.data["4. close"].min()
max_value = stock.data["4. close"].max()
min_value, max_value
```

```python
difference = max_value - min_value
difference
```

```python
ratios = [0, 0.236, 0.382, 0.618, 1]
for ratio in ratios:
    print(round(max_value - difference * ratio, 2))
```
